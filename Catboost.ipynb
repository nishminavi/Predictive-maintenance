```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from catboost import CatBoostClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_curve, auc, ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt
import shap

# Mount Google Drive (if using Colab)
from google.colab import drive
drive.mount('/content/drive')

# Paths and constants
train_base_path = '/content/drive/MyDrive/Aerospace/train data'
test_base_path = '/content/drive/MyDrive/Aerospace/test data'
datasets = ['FD001', 'FD002', 'FD003', 'FD004']
column_names = ['unit', 'cycle'] + [f'op_setting_{i}' for i in range(1, 4)] + [f'sensor_{i}' for i in range(1, 22)]
RUL_THRESHOLD = 30

# Helper functions
def load_and_clean_data(path):
    df = pd.read_csv(path, delim_whitespace=True, header=None, names=column_names)
    df = df.drop_duplicates().sort_values(by=['unit', 'cycle']).reset_index(drop=True)
    low_var_cols = [col for col in df.columns if df[col].nunique() <= 1]
    df.drop(columns=low_var_cols, inplace=True)
    return df

def calculate_rul(df):
    max_cycle = df.groupby('unit')['cycle'].max().reset_index()
    max_cycle.columns = ['unit', 'max_cycle']
    df = df.merge(max_cycle, on='unit', how='left')
    df['RUL'] = df['max_cycle'] - df['cycle']
    df.drop('max_cycle', axis=1, inplace=True)
    return df

def create_binary_labels(df, threshold=RUL_THRESHOLD):
    df['label'] = df['RUL'].apply(lambda x: 1 if x <= threshold else 0)
    return df

def plot_conf_matrix(y_true, y_pred, dataset_name):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Healthy", "Failing"])
    disp.plot(cmap='Blues')
    plt.title(f"📊 Confusion Matrix — {dataset_name}")
    plt.show()

def plot_roc_curve(y_true, y_probs, dataset_name):
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}', color='darkorange')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.title(f"🧪 ROC Curve — {dataset_name}")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()

# Dicts to hold results for comparison plots
results = {}
global_shap_dict = {}

for name in datasets:
    print(f"\n🚀 Processing Dataset: {name}")

    # Load and prepare train data
    df_train = load_and_clean_data(os.path.join(train_base_path, f'train_{name}.txt'))
    df_train = calculate_rul(df_train)
    df_train = create_binary_labels(df_train)
    feature_cols = df_train.columns.difference(['unit', 'cycle', 'RUL', 'label'])

    scaler = MinMaxScaler()
    df_train[feature_cols] = scaler.fit_transform(df_train[feature_cols])
    X_train = df_train[feature_cols]
    y_train = df_train['label']

    # Train CatBoost model
    model = CatBoostClassifier(verbose=0, random_state=42)
    model.fit(X_train, y_train)
    print("✅ Model trained.")

    # Load and prepare test data
    df_test = load_and_clean_data(os.path.join(test_base_path, f'test_{name}.txt'))
    df_test = calculate_rul(df_test)
    df_test = create_binary_labels(df_test)
    df_test[feature_cols] = scaler.transform(df_test[feature_cols])
    X_test = df_test[feature_cols]
    y_test = df_test['label']

    # Predict & evaluate
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    print("📈 Classification Report:")
    print(classification_report(y_test, y_pred, target_names=["Healthy", "Failing"]))

    # SHAP explanation
    X_sample = X_test.sample(n=min(100, len(X_test)), random_state=42)
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X_sample)

    # Mean absolute SHAP per feature
    shap_abs = np.abs(shap_values).mean(axis=0)
    global_shap_dict[name] = pd.Series(shap_abs, index=X_sample.columns)

    # SHAP summary plots
    shap.summary_plot(shap_values, X_sample, plot_type="bar", show=True)
    shap.summary_plot(shap_values, X_sample, show=True)

    # Plots
    plot_conf_matrix(y_test, y_pred, name)
    plot_roc_curve(y_test, y_prob, name)

    # Save results for combined plots
    results[name] = {
        'y_test': y_test,
        'y_pred': y_pred,
        'y_prob': y_prob
    }

# Global SHAP comparison plot
global_shap_df = pd.DataFrame(global_shap_dict).fillna(0)
plt.figure(figsize=(14, 8))
global_shap_df.T.plot(kind='bar', width=0.85)
plt.title("Global SHAP Feature Importance Comparison")
plt.ylabel("Mean Absolute SHAP Value")
plt.xlabel("Dataset")
plt.xticks(rotation=0)
plt.grid(True, axis='y')
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Combined ROC curves
plt.figure(figsize=(8, 6))
for name in datasets:
    y_test = results[name]['y_test']
    y_prob = results[name]['y_prob']
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.title("ROC Curve Comparison — All Datasets")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.grid(True)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# Combined confusion matrix grid
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.ravel()

for i, name in enumerate(datasets):
    y_test = results[name]['y_test']
    y_pred = results[name]['y_pred']
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(cm, display_labels=["Healthy", "Failing"])
    disp.plot(ax=axs[i], cmap='Blues', colorbar=False)
    axs[i].set_title(f"{name}")

plt.suptitle("Confusion Matrices — All Datasets", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import os
import pandas as pd

# Mount Google Drive (in case you haven't already)
from google.colab import drive
drive.mount('/content/drive')

# Paths
train_base_path = '/content/drive/MyDrive/Aerospace/train data'
datasets = ['FD001', 'FD002', 'FD003', 'FD004']
column_names = ['unit', 'cycle'] + \
               [f'op_setting_{i}' for i in range(1, 4)] + \
               [f'sensor_{i}' for i in range(1, 22)]
RUL_THRESHOLD = 30

# Functions
def load_and_clean_data(path):
    df = pd.read_csv(path, delim_whitespace=True, header=None, names=column_names)
    df = df.drop_duplicates().sort_values(by=['unit', 'cycle']).reset_index(drop=True)
    low_var_cols = [col for col in df.columns if df[col].nunique() <= 1]
    df.drop(columns=low_var_cols, inplace=True)
    return df

def calculate_rul(df):
    max_cycle = df.groupby('unit')['cycle'].max().reset_index()
    max_cycle.columns = ['unit', 'max_cycle']
    df = df.merge(max_cycle, on='unit', how='left')
    df['RUL'] = df['max_cycle'] - df['cycle']
    df.drop(columns=['max_cycle'], inplace=True)
    return df

def create_binary_labels(df, threshold=30):
    df['label'] = df['RUL'].apply(lambda x: 1 if x <= threshold else 0)
    return df

# 🔧 Create the rul_labeled_data dictionary
rul_labeled_data = {}

for name in datasets:
    print(f"🔄 Loading and labeling dataset: {name}")
    df = load_and_clean_data(os.path.join(train_base_path, f'train_{name}.txt'))
    df = calculate_rul(df)
    df = create_binary_labels(df, threshold=RUL_THRESHOLD)
    rul_labeled_data[name] = df
    print(f"✅ {name} loaded. Shape: {df.shape}")
```

```python
#Enhanced CatBoost

import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, ConfusionMatrixDisplay, auc
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE
from catboost import CatBoostClassifier
import matplotlib.pyplot as plt
import seaborn as sns
import shap

# Store results and global SHAP values
results = {}
global_shap_dict = {}

# Iterate over each dataset
for name in ['FD001', 'FD002', 'FD003', 'FD004']:
    print(f"\n🔄 Processing dataset: {name}")

    # Get labeled DataFrame
    df = rul_labeled_data[name].copy()

    # Drop non-feature columns
    X = df.drop(columns=['unit', 'cycle', 'RUL', 'label'])
    y = df['label']

    # Normalize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Apply SMOTE for balancing
    sm = SMOTE(random_state=42)
    X_resampled, y_resampled = sm.fit_resample(X_scaled, y)

    # Split into training and validation sets
    X_train, X_val, y_train, y_val = train_test_split(
        X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
    )

    # Hyperparameter tuning (basic grid)
    param_grid = {
        'depth': [4, 6],
        'learning_rate': [0.05, 0.1],
        'iterations': [300, 500]
    }

    cat_model = CatBoostClassifier(verbose=0, random_state=42)
    grid = GridSearchCV(cat_model, param_grid, cv=3, scoring='f1', n_jobs=-1)
    grid.fit(X_train, y_train)

    # Best model from tuning
    best_model = grid.best_estimator_

    # Predictions
    y_pred = best_model.predict(X_val)
    y_prob = best_model.predict_proba(X_val)[:, 1]

    # Metrics
    accuracy = accuracy_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred)
    roc_auc_score_value = roc_auc_score(y_val, y_prob)

    print(f"✅ {name} — Accuracy: {accuracy:.4f}, F1: {f1:.4f}, AUC: {roc_auc_score_value:.4f}")

    # Save results
    results[name] = {
        'model': best_model,
        'accuracy': accuracy,
        'f1': f1,
        'auc': roc_auc_score_value,
        'X_val': X_val,
        'y_val': y_val,
        'y_pred': y_pred,
        'y_prob': y_prob
    }

    # Confusion Matrix
    cm = confusion_matrix(y_val, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Healthy', 'Fail'], yticklabels=['Healthy', 'Fail'])
    plt.title(f'Confusion Matrix — {name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_val, y_prob)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f'AUC = {roc_auc_score_value:.2f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f'ROC Curve — {name}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid(True)
    plt.show()

    # SHAP explanation
    print(f"🔍 Calculating SHAP values for {name}...")
    explainer = shap.TreeExplainer(best_model)
    shap_values = explainer.shap_values(X_val)

    shap_abs_mean = np.abs(shap_values).mean(axis=0)
    global_shap_dict[name] = pd.Series(shap_abs_mean, index=X.columns)

    # Summary plots (bar and beeswarm)
    shap.summary_plot(shap_values, X_val, feature_names=X.columns, plot_type="bar", show=True)
    shap.summary_plot(shap_values, X_val, feature_names=X.columns, show=True)

# === Global SHAP Feature Importance Comparison ===
print("\n📊 Global SHAP Feature Importance Comparison")
global_shap_df = pd.DataFrame(global_shap_dict).fillna(0)

plt.figure(figsize=(14, 8))
global_shap_df.T.plot(kind='bar', figsize=(14, 6), width=0.85)
plt.title("Global SHAP Feature Importance Comparison (All Datasets)", fontsize=16)
plt.ylabel("Mean Absolute SHAP Value")
plt.xlabel("Dataset")
plt.xticks(rotation=0)
plt.grid(True, axis='y')
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# === Combined ROC Curve Comparison ===
print("\n📊 Combined ROC Curve Comparison")

plt.figure(figsize=(8, 6))
for name in ['FD001', 'FD002', 'FD003', 'FD004']:
    y_val = results[name]['y_val']
    y_prob = results[name]['y_prob']
    fpr, tpr, _ = roc_curve(y_val, y_prob)
    roc_auc_score_value = auc(fpr, tpr) # Use sklearn.metrics.auc
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc_score_value:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.title("ROC Curve Comparison — All Datasets")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.grid(True)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# === Combined Confusion Matrix Grid ===
print("\n📊 Combined Confusion Matrix Grid")

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.ravel()

for i, name in enumerate(['FD001', 'FD002', 'FD003', 'FD004']):
    y_val = results[name]['y_val']
    y_pred = results[name]['y_pred']
    cm = confusion_matrix(y_val, y_pred)
    disp = ConfusionMatrixDisplay(cm, display_labels=["Healthy", "Fail"])
    disp.plot(ax=axs[i], cmap='Blues', colorbar=False)
    axs[i].set_title(f"{name}")

plt.suptitle("Confusion Matrices — All Datasets", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
